## Our aim is to implement classification of faces using PCA. To understand PCA better, we will implement it from scratch.

### Import all necessary packages

```python
import numpy as np
from sklearn.datasets import fetch_lfw_people
import matplotlib.pyplot as plt
```

###  Load the lfw_people dataset from sklearn. Retain pictures of those people that have atleast 200 faces 

```python
faces = fetch_lfw_people(min_faces_per_person=200)
print(faces.target_names)
print(faces.images.shape)
```

### Reshape the faces to a 2-d array. (Flattening)

```python
faces_data = faces.data
print(faces_data.shape)
```

## Now, we will try to implement PCA without using inbuilt function from sklearn

1. For that first, find the covariance matrix
$ \sigma  = \sum_{i=1}^{i=N}\frac{1}{N-1}(X_i-\mu)^T(X_i-\mu)$ where $\mu = \frac{1}{N} \sum_{i=1}^{i=N}{X_i}$



```python
mean = np.mean(faces_data,axis=1).reshape(-1,1)
faces_data_centered = np.subtract(faces_data,mean)
N = faces_data.shape[0]
cov_mat = np.matmul(np.transpose(faces_data_centered),faces_data_centered) / (N-1)

print(cov_mat.shape)
print(mean.shape)
```

### Do Eigen Value Decomposition

```python
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
abs_eig_vals = np.absolute(eig_vals)
```

### For selecting the number of principal components, plot the expected variance graph (only cumulative variance plot required) and choose minimum k which gives energy above 95%.

```python
tot = sum(abs_eig_vals)
var_exp = [(i / tot)*100 for i in sorted(abs_eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
```

```python
plt.plot(cum_var_exp)
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
plt.grid()
```

### Now we will sort the eigen vectors in the decreasing order of the eigen values. Use np.argsort() to get the indices of the sorted eigen values in ascending order. Flip it using np.flip() to get it in descending order. Now sort the eigen vectors in this order and then take the k components.

```python
# taking k as 135
ind = np.argsort(eig_vals)

asc_eig_vecs = eig_vecs[ind]
desc_eig_vecs = np.flip(eig_vecs)

new_eig_vecs = desc_eig_vecs[:,:135]
```

```python
downsampled_faces = np.matmul(faces_data,new_eig_vecs)
target = faces.target
```

### Do classification on the projected data and report the classification score

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split

trainX, testX, trainY, testY = train_test_split(downsampled_faces, target, train_size=0.75, random_state=0)

clf = LogisticRegression(C=0.001)
clf.fit(trainX,trainY)

pred= clf.predict(testX)

print("Train Score for Logistic Regression : {:.2f}".format(clf.score(trainX,trainY)))
print("Test Score for Logistic Regression : {:.2f}".format(clf.score(testX,testY)))
```
